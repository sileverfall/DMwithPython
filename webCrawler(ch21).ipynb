## Requests

```python
import requests

url = 'http://www.grandtech.info'
htmlfile = requests.get(url)
type(htmlfile) #回傳resoonse物件
```

```python
if htmlfile.status_code == requests.codes.ok:
    print("Successiful!")
else:
    print("Fail!")
```
